```python
import data_preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import folium
#import geopandas as gpd
import json

cities_short = ['ATL', 'BAL', 'BOS', 'CHAR', 'CHI', 'CLE', 'DEN', 'DET', 'HOU', 'IND', 'LV', 'LA', 'MIA', 'NASH', 'NO',
                'NYC', 'PHI', 'SF', 'SEA', 'DC']
VADER = data_preprocessing.get_mean_VADER()
SES_demographics = data_preprocessing.get_SES_demographics()
covid = data_preprocessing.get_mean_COVID_cases()
data = data_preprocessing.get_full_data()
FIPS_to_county = pd.read_csv('urban_rural/data/urban_counties.tsv', sep='\t')
map_data = pd.merge(data, FIPS_to_county[['FIPS', 'city']], left_on=data['NAME'].str[:5], right_on=FIPS_to_county['city'].str[:5], how='outer')
map_data.drop(['key_0','city'], axis=1, inplace=True)
map_data
```

```python
with open('us_2020_counties.json') as f:
    geo = json.load(f)

filtered_data = []
for item in geo['features']:
    if item['properties']['GEOID'] in map_data['FIPS']:
        filtered_data.append(item)

with open('filtered_data.json', 'w') as f:
    json.dump(filtered_data, f)

red = 'YlOrRd'
blue = 'BuPu'
green = 'YlGn'
def create_map(feature_name, color):
    # create a map centered on the USA
    usa_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

    # add choropleth layer to the map
    folium.Choropleth(
        geo_data='us_2020_counties.json',
        name='choropleth',
        data=data,
        columns=['NAME', feature_name],
        key_on='feature.properties.NAME',
        fill_color=color,
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name=feature_name
    ).add_to(usa_map)

    # add layer control to the map
    folium.LayerControl().add_to(usa_map)

    # display the map
    return usa_map


map = create_map('MEDIAN_AGE', red)
map
```
